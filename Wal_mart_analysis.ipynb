```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import datetime as dt
import prophet as ppt
import warnings
warnings.filterwarnings ("ignore")
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

### ***Obtenci√≥n de Datos***



```python
#se cargan los datos de entrada proporcionado en el proyecto de las tiendas Wal-mart, 
#asimismo se hace la conexi√≥n de los datos a Github mostrado en la siguiente ruta
#de la misma forma se indica a Python elija los primeros 20 registros
url = 'https://raw.githubusercontent.com/Psyclophe/Data-Scientist/main/Walmart_Store_sales.csv'
df = pd.read_csv (url)
df.head(20)
```

```python
#con esta funcion se da la instrucci√≥n a Python para que nos muestre los datos y, poder detectar la cantidad de datos nulos
df.isnull().sum()
```

```python
#econ la carga de informaci√≥n y con esta funcion se pide a Python indicar el tipo de dato de cada columna, 
#para analizar si es necesario hacer cambios en el tipo de dato en el dataframe
#se identifico que "Date" es un dato que es objeto y se debe convertir a dato fecha
df.info()
```

```python
#en esta funci√≥n se pide a Python convertir los datos objeto a string
#aqu√≠ podemos visualizar que el dato que era un objeto "Date" se conviritio al formato fecha en donde se visualiza "datetime64[ns]"
df["Date"]=pd.to_datetime(df["Date"])
df.info()
```

```python
#con esta funcion se da la instrucci√≥n a Python para ver las medidas de tendencia central de cada columna
df.describe()
```

```python
#se da la instruccion a Python que se ordene las fechas de menor a mayor en el archivo que se va a analizar
#con esta funcion se le indica al programa que ordene por fechas, de la fecha mas antigua hasta la fecha mas reciente
data_sort=df.sort_values("Date")
data_sort
```

```python
#con la funcion groupby se inidca al sistema sumar y agrupar por tienda y por semana
#al ejecutar la funcion, el programa ordena por tienda y agrega la columna "0" a la columna "1"
group=df.groupby("Store").sum("Weekly_Sales")
group.reset_index(level=0, inplace=True)
group
```

```python
std = df.groupby('Store')['Weekly_Sales'].std()
std = pd.DataFrame(std)
std_sort=std.sort_values('Weekly_Sales', ascending=False)
std_sort
```

```python
sort=group.sort_values('Weekly_Sales')
sort
```

```python
store1 = df[df["Store"]==1]
store1 = store1.sort_values('Date')
store1
```

```python
#con esta funcion se pide a Python identificar los d√≠as festivos
df[df["Holiday_Flag"]==1]
```
