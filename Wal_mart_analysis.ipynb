```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings ("ignore")
import datetime as dt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#cargamos los datos de entrada de las tiendas de Wal-mart, y se indica a Python elija los primeros 20 registros
#asimismo se hace la conexcion de los datos en Github
url = 'https://raw.githubusercontent.com/Psyclophe/Data-Scientist/main/Walmart_Store_sales.csv'
df = pd.read_csv (url)
df.head(20)
```

```python
#en esta funcion podemos visualizar en Python el tipo de dato de cada columna, para analizar si es necesario hacer cambios en el tipo de dato en el dataframe
#se identifico que "Date" es un objeto y se debe convertir a dato fecha
df.info()
```

```python
#con esta funcion se da la instrucción a Python para poder ver las medidas de tendencia central de cada columna
df.describe()
```

```python
#con esta funcion se da la instrucción a Python para que nos muestre los datos y, poder detectar la cantidad de datos nulos
df.isnull().sum()
```

```python
#en esta linea se pide a Python convertir los datos objeto a string
#aquí podemos visualizar que el dato que era un objeto "Date" se conviritio al formato fecha en donde se visualiza "datetime64[ns]"
df["Date"]=pd.to_datetime(df["Date"])
df.info()
```

```python
#se da la instruccion a Python que se ordene las fechas de menor a mayor en el archivo que se va a analizar
#con esta funcion se le indica al programa que ordene por fechas, de la fecha mas antigua hasta la fecha mas reciente
data_sort=df.sort_values("Date")
data_sort
```

```python
#con la funcion groupby se inidca al sistema agrupar por tienda y por semana
#al ejecutar la funcion, el programa ordena por tienda y agrega la columna "0" a la columna "1"
group=df.groupby("Store").sum("Weekly_Sales")
group.reset_index(level=0, inplace=True)
group
```

```python
std = df.groupby('Store')['Weekly_Sales'].std()
std = pd.DataFrame(std)
std_sort=std.sort_values('Weekly_Sales', ascending=False)
std_sort
```

```python
sort=group.sort_values('Weekly_Sales')
sort
```

```python
store1 = df[df["Store"]==1]
store1 = store1.sort_values('Date')
store1
```

```python
fig = px.box(data_sort, y = "Weekly_Sales", x = "Store",  color = "Store")
fig.show()
```

```python
sort5=sort[['Weekly_Sales', 'Store']][-5:]
sort5=sort5.sort_values('Weekly_Sales')
sort5
```

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(30,12))
sns.lineplot(data=store1, x='Date', y='Weekly_Sales', hue='Store', palette='tab10')
plt.show()
```

```python
f, ax=plt.subplots(figsize=(12,8))
sns.barplot(x='Store', y='Weekly_Sales', data=sort5, order=sort5.sort_values('Weekly_Sales').Store)
plt.show()
```
